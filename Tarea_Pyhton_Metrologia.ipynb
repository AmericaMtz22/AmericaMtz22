--> count : número de datos
--> mean : media aritmetica
--> std : desviación estandar
--> min : dato minimo encontrado

--> max : valor maximo encontrado

```
# Con eso en cuenta, se utiliza la libreria de statistics para realizar la comparación. Debido a que no todos los datos que nos da pandas se encuentran en la libreria de statistics, se utilizan en conjunto las funciones predeterminadas de python.
```



````python
#Función para calcular percentiles
def percentile(data, percentile):
    data_sorted = sorted(data)
    n = len(data_sorted)
    index = (percentile / 100) * (n - 1)
    if index.is_integer():
        return data_sorted[int(index)]
    else:
        lower_index = int(index // 1)
        upper_index = lower_index + 1
        lower_value = data_sorted[lower_index]
        upper_value = data_sorted[upper_index]
        fraction = index % 1
        return lower_value + fraction * (upper_value - lower_value)
````
